```python
import os

os.environ["HUGGINGFACEHUB_API_TOKEN"] = ""
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint


llm = HuggingFaceEndpoint(
     model="deepseek-ai/DeepSeek-V3.1",
     provider="auto"
)

chat_model = ChatHuggingFace(llm=llm)
```

```python
help(HuggingFaceEndpoint)
```

```python
from re import template
from langchain.messages import (
    HumanMessage,
    SystemMessage,
)
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variable = "game_name",
    template = "Give me name of teams {game_name} world wordcup wining team from starting with captain name, country and year in a tabular structure"
)
user_query = prompt.format(game_name="footbal")

messages = [
    SystemMessage(content="You're a helpful assistant to human"),
    HumanMessage(
        content=user_query
    ),
]

chat_model.invoke(messages).pretty_print()
```
